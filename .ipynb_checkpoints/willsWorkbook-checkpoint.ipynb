```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv("housing.csv")
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['total_bedrooms'].fillna(0,inplace = True)
```

```python
df.corr()
```

```python
df.describe()
```

```python



plt.figure(figsize=(10,10))
plt.subplots_adjust(hspace = .25)
plt.subplot(1,1,1)
plt.xlabel('ocean_proximity',fontsize=12)
plt.ylabel('median_house_value',fontsize=12)
sns.boxplot(data=df,x='ocean_proximity',y='median_house_value')

#sns.catplot(x="total_bedrooms", y="median_house_value", hue="housing_median_age",
           # kind="violin", data=df);


plt.plot()
```

```python
plt.figure(figsize=(55,10))
plt.subplots_adjust(hspace = .25)
plt.subplot(1,2,2)
plt.xlabel('Housing Median Age',fontsize=16)
plt.ylabel('Median House Value',fontsize=16)
sns.boxplot(data=df,x='housing_median_age',y='median_house_value')
plt.plot()
```
