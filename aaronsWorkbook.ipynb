```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("housing.csv")
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['total_bedrooms'].fillna(0,inplace = True)
```

```python
df.corr()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,6))
sns.distplot(df['housing_median_age'],color='blue')
plt.xlabel('Housing Median Age', fontsize=14)
plt.show()
```

```python
df[df['housing_median_age']>30]['housing_median_age'].value_counts().head
```

```python
df=df.loc[df['housing_median_age']<36,:]
df=df[df['population']<25000]
plt.figure(figsize=(12,6))
sns.distplot(df['housing_median_age'],color='green')
plt.xlabel('Housing Median Age', fontsize=14)
plt.show()
```
